```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

import sys
sys.path.append("Common_tools")
from data_visualization import img_combine
```

```python
data_dir = "/data/seanyu/UPMC_pathology/Pathology_Analysis/data/"
label_dir = "/data/seanyu/UPMC_pathology/Pathology_Analysis/splits/sp0/"

def get_image(img_path):
    im = cv2.imread(img_path)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    return im
```

```python
df_view = pd.read_csv(label_dir + "tr_lst", header=None)
df_view.columns = ['pid', 'cate']
df_view['img_path'] = df_view['pid'].apply(lambda x: os.path.join(data_dir, 'image' + str(x) + '.png'))

df_cate0 = df_view[df_view['cate'] == 0]
df_cate1 = df_view[df_view['cate'] == 1]
df_cate0.reset_index(drop = True, inplace=True)
df_cate1.reset_index(drop = True, inplace=True)
```

```python
# View first 100 images of category 0
arrays = np.array([get_image(i) for i in df_cate0.iloc[0:100].img_path])
img_combine(arrays, ncols=10, size=2)
```

```python
# View first 100 images of category 1
arrays = np.array([get_image(i) for i in df_cate1.iloc[0:100].img_path])
img_combine(arrays, ncols=10, size=2)
```
